# Visualization 

##### Scatter plot for numerical data relationship analysis

```python
plt.scatter(new1_df['Global_active_power'],new1_df['Voltage'])
plt.title('Global_active_power vs Voltage')
plt.xlabel('Global_active_power')
plt.ylabel('Voltage')
plt.xticks([])
plt.show()
```

##### Histogram for freequency distribution of numerical data

```python
plt.hist(new1_df['Global_active_power'],color='green',edgecolor='yellow',bins = 5)
plt.title('Global_active_power Histogram')
plt.xlabel('Global_active_power')
plt.ylabel('Freequecy')
plt.show()
```

##### Barplot for freequency distribution of categorical data

```python
loc_names = new1_df['location'].unique()
print(loc_names)
```

```python
loc_num = len(new1_df['location'].value_counts())
loc_range = np.arange(loc_num)
loc_range
```

```python
loc_valuecounts = new1_df['location'].value_counts()

#loc_valuecounts=list(new1_df['location'].value_counts())
loc_valuecounts
```

```python
index = loc_range
freq = loc_valuecounts
category = loc_names


plt.bar(index,freq,color=['r','b','g'])
plt.title('Bar plot for Location based intensity')
plt.xlabel('Locations')
plt.ylabel('freequency')
plt.xticks(index,category,rotation =90)
plt.show()

```

another example for bar chart:

```python
print(loc_names)

freq = [300,500,600]
print(freq)

index = np.arange(len(loc_names))
print(index)
```

##### scatterplot of 2 variables based on another avriable - lmplot


```python
#use sns.lmplot - combination of regression plot + FacetGrid
#used when doing conditional subset of data or when using 3 variables
```

```python
sns.lmplot(
    x = 'Global_active_power',
    y = 'Voltage',
    data = new1_df,
    hue = 'location',
    fit_reg = False,
    legend = True,
    palette = 'Set2',
    markers=["p", "x","*"]
    )
plt.show()
```
