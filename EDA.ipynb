# Freequency tables

### pandas.crosstab - Computes freequency table of one,two or more factors

##### freequency distribution of 1 factor

```python
pd.crosstab(index = newdf['Team'],columns = "count",dropna=True)
```

##### Tow way table -freequency distribution of 2 factors

```python
pd.crosstab(index = df['Team'],columns = df['Gender'],dropna=True)
```

##### Two-way joint probabilty  - gives probability value 

```python
pd.crosstab(index =df['Team'],columns =df['Gender'],normalize=True,dropna=True)
```

```python
##### Two-way marginal probabilty  - gives probability value based on a single value -gets row sums and column sums
```

```python
pd.crosstab(index =df['Team'],columns =df['Gender'],normalize=True,margins=True,dropna=True)
```

```python
##### Two-way Conditional probabilty
```

```python
pd.crosstab(index = df['Team'],columns=df['Gender'],normalize='index',dropna=True,margins=True)

# Given the Teams, what is the probabilty of Female and Male.
#Here sum of rows =1
```

```python
#Given the Gender of Employees, the probability of a teams
# here sum of columns =1

pd.crosstab(index = df['Team'],columns=df['Gender'],normalize='columns',dropna=True,margins=True)
```
