```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter('ignore')
```

```python
x = np.arange(1,11)
z = np.random.randint(1,100,10)
y = x*2
x,y,z
```

```python
plt.plot(x,z,color = 'cyan',linestyle = ':',linewidth=2)
plt.title("Non-linear graph")
plt.xlabel("x-value")
plt.ylabel("z-value")
plt.show()
```

```python
plt.plot(x,y)
plt.title("Linear graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
plt.plot(x,y,color = 'g',linestyle = ':',linewidth = 3)
plt.title("Linear graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
plt.plot(x,y,color='hotpink',linestyle = ':',linewidth = '4')
plt.plot(x,z,color = 'cyan',linestyle='-.',linewidth ='3')
plt.title("Not so great graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
x = np.random.randint(1,10,5)
y = x*2
z = x*3

plt.plot(x,y,color = 'red')
plt.plot(x,z,color = 'green')
plt.grid(True)
plt.show()
```

```python
x = np.random.randint(1,10,5)
y = x*2
z = x*3

plt.subplot(2,2,1)
plt.plot(x,y,color = 'red')

plt.subplot(2,2,2)
plt.plot(x,z,color = 'green')

plt.subplot(2,2,3)
plt.plot(x,y,color = 'hotpink')

plt.subplot(2,2,4)
plt.plot(x,z,color = 'cyan')
```

```python
marks = {'A':98,'B':55,'C':75,'D':30,'E':80}
names = list(marks.keys())
values = list(marks.values())
values
```

```python
plt.bar(names,values)
plt.title("Mark distribution")
plt.xlabel("Names")
plt.ylabel("Marks Obtained")
plt.show()
```

```python
plt.barh(names,values,color = 'g')
plt.title("Mark distribution")
plt.ylabel("Names")
plt.xlabel("Marks Obtained")
plt.grid(True)
plt.show()
```
