```python
import pandas as pd
```

### Series

```python
s1 = pd.Series([20,30,40,50,60]) # series is a LABELLED 1-dimentional array(inside is a list)
s1

type(s1)
```

```python
list1 = ['apple','banana','grapes'] # using index argument for list
fruitseries = pd.Series(list1)
fruitseries
fruitseries = pd.Series(list1,index = [11,12,13])
fruitseries
```

```python
alphabet = pd.Series({'a':'Apple','b':'Banana','c':'Cat','d':'Doll'}) # passing a dictionary
alphabet

dict1 = {'dog':'Domestic animal','lion':'Wild animal','elephant':'Wild but harmless'}
animalSeries = pd.Series(dict1)
animalSeries
```

```python
dictz ={'k1':'abc','k2':'hgf','k3':'klt'}
dS1 = pd.Series(dictz) # dictionary in Series
dS1
dS2 = pd.Series(dictz,index = ['k2','k1','k3']) # change position of the key values using index
dS2
dS3 = pd.Series(dictz,index = ['k2','k1','k3','k4']) # a new index value k4 is add but the result will be NaN
dS3
```

```python
e = pd.Series(['apple','banana','grapes','orange','strawberry','kiwi','mango','jackfruit','peach','berry']) 

e[0]# extracting value item in series
e[3]
e[:3]
e[1:4]
e[-3:]
e[-4:-2]
```

```python
s1 = pd.Series([20,30,40,50,60]) #scalar operatons in series
s1+5
s1/10
s1-10
s1*5
s2 = pd.Series([2,3,4,5,6])
s1-s2
```

```python
# apply any functions using .apply(function name)
def anyfunc(x):
    return x*10

gg = iris.apply(anyfunc)   # for the entire dataframe
gg
h = iris[['SepalLengthCm','PetalWidthCm']].apply(anyfunc) # for the specific columns
h


def half(s):
    return s/2

iris[['SepalLengthCm','PetalWidthCm']].apply(half)
```
